```python
import spacy
import re
from pythainlp.tokenize import word_tokenize
import pytextrank

# Load spaCy's small English model
nlp = spacy.load("en_core_web_sm")

# Add PyTextRank to spaCy pipeline
nlp.add_pipe("textrank")

# Tokenize the Thai text using PyThaiNLP
tokens = word_tokenize(check, custom_dict=custom_trie, engine="newmm")

# Join the tokens into a processed string for spaCy
processed_text = " ".join(tokens)

# Process the tokenized text with spaCy
doc = nlp(processed_text)

# Load Thai stopwords

# Define custom stopwords if needed
custom_stopwords = ['คำที่ไม่ต้องการ']  # Example, add your custom stopwords here

# Initialize a list to store keywords with scores
keywords_with_scores = []

# Extract the top-ranked phrases
for phrase in doc._.phrases:  # Iterate through phrases to get scores
    for chunk in phrase.chunks:  # Iterate through words in each phrase
        keyword = chunk.text
        # Filter out Thai stop words, custom stop words, but keep English alphabets
        filtered_keyword_words = [
            word for word in keyword.split()  # Split the phrase into words
            if word not in thai_stopwords() and word not in custom_stopwords
            and not re.search(r'\d', word)  # Exclude words with numbers
            or re.search('[a-zA-Z]', word)  # Keep words with alphabetic characters
        ]
        filtered_keyword = " ".join(filtered_keyword_words)
        
        # Append the filtered keyword if it's not already in the list
        if filtered_keyword and filtered_keyword not in [kw[0] for kw in keywords_with_scores]:
            keywords_with_scores.append((filtered_keyword))

# Print the extracted keywords
print(f"Keywords: {keywords_with_scores}")
```
