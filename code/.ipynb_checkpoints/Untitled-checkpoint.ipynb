```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForMaskedLM
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_abstract.csv')
df.head()
```

```python
text = df['content'][0]
```

```python
tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")
model = AutoModelForMaskedLM.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")

# Tokenize the text
inputs = tokenizer(text, return_tensors="pt")

# Display tokenized inputs
print("Tokenized Inputs:", inputs)

# If you want to check the tokens (word pieces), you can use:
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)
```
