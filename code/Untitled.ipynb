### Streamlit

```python
import streamlit as st
import pandas as pd
import datetime

# ฐานข้อมูลสมมติสำหรับเก็บข้อมูล Traffic
if 'user_logs' not in st.session_state:
    st.session_state.user_logs = []

# บันทึกการคลิก
def log_interaction(document_id):
    log = {
        'timestamp': datetime.datetime.now(),
        'document_id': document_id,
        'action': 'click'
    }
    st.session_state.user_logs.append(log)
    st.success(f"Logged interaction for document {document_id}")

# UI แสดงเอกสาร
st.title("Document Recommender System")
documents = [{'id': 1, 'title': 'โลจิสติกส์เบื้องต้น'}, {'id': 2, 'title': 'นวัตกรรมในโลจิสติกส์'}]

for doc in documents:
    st.write(doc['title'])
    if st.button(f"View {doc['title']}", key=doc['id']):
        log_interaction(doc['id'])

# บันทึก Traffic
if st.button("Save Logs"):
    pd.DataFrame(st.session_state.user_logs).to_csv("user_logs.csv", index=False)
    st.success("Saved logs to user_logs.csv")
```

```python
import numpy as np

# คะแนนความนิยม (Popularity Score)
popularity_scores = {'doc1': 0.8, 'doc2': 0.4, 'doc3': 0.6}

# คะแนนความคล้ายคลึง (Content Similarity)
content_similarity_scores = {'doc1': 0.7, 'doc2': 0.9, 'doc3': 0.5}

# Hybrid Recommendation: ผสมคะแนน
hybrid_scores = {doc: 0.5 * popularity_scores[doc] + 0.5 * content_similarity_scores[doc]
                 for doc in popularity_scores.keys()}

# เรียงลำดับเอกสารที่แนะนำ
recommended_docs = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)
st.write("Recommended Documents:", recommended_docs)
```
